```python
import numpy as np
import cv2 as cv
```

```python
## Code 1
while True:
    frame = cv.imread("smarties.png")
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    l_b = np.array([110,50,50])
    l_u = np.array([130,255,255])
    
    mask = cv.inRange(hsv, l_b, l_u)
    
    res = cv.bitwise_and(frame, frame, mask=mask)
    
    cv.imshow("Frame", frame)
    cv.imshow("Mask", mask)
    cv.imshow("Res", res)
    
    if (cv.waitKey(1) == 27):
        break
    
cv.destroyAllWindows()
```
